```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv("./dataset/movie/tmdb_5000_movies.csv")
credits = pd.read_csv("./dataset/movie/tmdb_5000_credits.csv")
```

```python
movies.head()
```

```python
credits.head()
```

```python
movies = movies.merge(credits, on="title")
```

```python
# genres
# movie_id
# keywords
# title
# overview
# release_date
# cast
# crew

movies = movies[["movie_id", "title", "overview", "genres", "keywords", "cast", "crew"]]
movies.head()
```

```python
# Start of preprocessing
```

```python
# remove missing data
movies.isnull().sum()
```

```python
# drop rows where data in not available.
# in current case drop rows where overview is not available
movies.dropna(inplace=True)
```

```python
# check if duplicate data exist
movies.duplicated().sum()

# in current case there are no duplicated columns but incase there exist duplicates then remove those duplicates
```

```python
# format json object columns.
# starting with genres column
movies.iloc[0]["genres"]
```

```python
# format can be done in following way
# existing column - {"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]
# formatted column - ["Action", "Adventure", "Fantasy", "Science Fiction"]

# there is one more problem that the existing column is a string column so we need to convert string to list of object (dict)
```

```python
import ast
ast.literal_eval(movies.iloc[0]["genres"])
```

```python
def formatData(obj):
    formattedList = []
    for i in ast.literal_eval(obj):
        formattedList.append(i["name"])
    return formattedList
```

```python
# format genres column
movies["genres"] = movies["genres"].apply(formatData)
```

```python
movies.head()
```

```python
movies["keywords"] = movies["keywords"].apply(formatData)
```

```python
movies.head()
```

```python
def formatLargeData(obj):
    formattedList = []
    counter = 0
    for i in ast.literal_eval(obj):
        counter+=1
        formattedList.append(i["name"])
        if counter == 3:
            break
    return formattedList
```

```python
movies["cast"] = movies["cast"].apply(formatLargeData)
```

```python
movies.head()
```

```python
def fetchDirector(obj):
    formattedList = []
    for i in ast.literal_eval(obj):
        if i["job"] == "Director":
            formattedList.append(i["name"])
            break
    return formattedList
```

```python
movies["crew"] = movies["crew"].apply(fetchDirector)
```

```python
movies.head()
```

```python
movies["overview"] = movies["overview"].apply(lambda x: x.split())
```

```python
movies.head()
```

```python
# The spaces that exist in between 2 words should be removed as it may cause problems when creating tags
# For eg: 'Sam Worthington' is name of 1 person but when we are creating tags we will end up with 2 tags one for 'sam' and 
# one for 'Worthington' and this may cause problems when recommeding movies to user.
# It may happen that user should be recommended movies of 'Sam Mendes' but we end up showing up movies of 'Sam Worthington' which is not what we want.
```

```python
movies["genres"] = movies["genres"].apply(lambda x: [i.replace(" ", "") for i in x])
movies["keywords"] = movies["keywords"].apply(lambda x: [i.replace(" ", "") for i in x])
movies["cast"] = movies["cast"].apply(lambda x: [i.replace(" ", "") for i in x])
movies["crew"] = movies["crew"].apply(lambda x: [i.replace(" ", "") for i in x])
```

```python
movies.head()
```

```python
# create new column called tags which will be concatination of overview, genres, keywords, cast and crew columns
movies["tags"] = movies["overview"] + movies["genres"] + movies["keywords"] + movies["cast"] + movies["crew"]
```

```python
movies.head()
```

```python
movie_df = movies[["movie_id", "title", "tags"]]
```

```python
movie_df.head()
```

```python
# convert list in tags column to string
movie_df['tags'] = movie_df["tags"].apply(lambda x: " ".join(x))
```

```python
# convert everything to lowercase
movie_df['tags'] = movie_df["tags"].apply(lambda x: x.lower())
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
movie_df['tags'] = movie_df['tags'].apply(stem)
```

```python
movie_df.head()
```

```python
# text vectorization
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
vectors = cv.fit_transform(movie_df['tags']).toarray()
```

```python
# names used for vectorization
cv.get_feature_names_out()
```

```python
vectors
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity
```

```python
sorted(list(enumerate(similarity[0])), reverse=True, key=lambda x:x[1])[1:6]
```

```python
def recommendMovie(movie):
    movie_index = movie_df[movie_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movie_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x:x[1])[1:6]

    for i in movie_list:
        print(movie_df.iloc[i[0]].title)
    return movie_index

recommendMovie('Batman Begins')
```
